```python
import matplotlib.pyplot as plt
import numpy as np
import os

os.makedirs("imgs", exist_ok=True)
```

## Line Plot - Weekly Temperature

This graph shows temperature changes during the week.
I chose a line plot because it is useful for showing trends over time.

```python
from matplotlib.typing import MarkerType
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temp = [18, 20, 19, 22, 23, 24, 21]

plt.figure(figsize=(8,5))
plt.plot(days, temp,
         marker="o",
         linestyle="--",
         color="blue",
         label="Temperature",)

plt.title("Weekly Temperature")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True)

plt.savefig("imgs/line_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Bar Chart - Student Scores

This graph compares the scores of different students.
I chose a bar chart because it is ideal for comparing categories.

```python
students = ["Ana", "Luis", "Maria", "John", "Emma"]
scores = [85, 92, 78, 88, 95]

plt.figure(figsize=(8,5))
plt.bar(students, scores,
        color="orange")

plt.title("Student Scores")
plt.xlabel("Students")
plt.ylabel("Score")

plt.savefig("imgs/bar_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Histogram - Exam Score Distribution

This histogram shows how exam scores are distributed.
I chose a histogram because it helps visualize frequency and distribution.

```python
np.random.seed(42)
scores = np.random.randint(10, 100, size=10)

plt.figure(figsize=(8,5))
plt.hist(scores,
         bins=10,
         color="green",
         edgecolor="black")

plt.title("Exam Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.savefig("imgs/histogram.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Scatter Plot - Study Hours vs Grades

This graph shows the relationship between study hours and grades.
I chose a scatter plot because it helps identify correlations between variables.

```python
study_hours = [1,2,3,4,5,6,7,8]
grades = [50,55,60,65,70,78,85,90]

plt.figure(figsize=(8,5))
plt.scatter(study_hours, grades,
            color="red",
            marker="x",
            s=100,
            label="Students")

plt.title("Study Hours vs Grades")
plt.xlabel("Study Hours")
plt.ylabel("Grades")
plt.legend()

plt.savefig("imgs/scatter_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Pie Chart - Monthly Expenses

This graph represents the percentage of monthly expenses by category.
I chose a pie chart because it clearly shows proportions of a whole.

```python
categories = ["Rent", "Food", "Transport", "Entertainment", "Other"]
expenses = [40, 25, 15, 10, 10]

plt.figure(figsize=(8,5))
plt.pie(expenses,
        labels=categories,
        autopct="%1.1f%%",
        startangle=90)

plt.title("Expense Distribution")

plt.savefig("imgs/pie_chart.png", dpi=300, bbox_inches="tight")
plt.show()
```
